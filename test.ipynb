```python
%load_ext autoreload
%autoreload 2

from physiological_signals import predicted_and_gt
from metrics import calculate_HR_metrics
import matplotlib.pyplot as plt
```

## Subject
- Subject index from 1 to 3

```python
sub_idx = '3' # '1', '2', '3'
```

## Load predicted PPG and ground truth PPG
 - Parameters
     - video_path : 얼굴 영상 (vid.avi) 이 저장된 경로
     - bvp_path : referenc PPG (ground_truth.txt) 가 저장된 경로 
 - Output
     - pred : predicted PPG
     - gt : ground truth PPG

```python
pred, gt = predicted_and_gt(video_path='./Data/sub' + sub_idx + '/vid.avi', bvp_path='./Data/sub' + sub_idx + '/ground_truth.txt')
```

## Figure

```python
fig, axes = plt.subplots(2, 1, figsize=(15, 8), gridspec_kw={'hspace': 0.5})

Start_point = 0
End_point = 500

#############################################################################
# Predicted PPG
axes[0].plot(pred[Start_point:End_point], 'dodgerblue')
axes[0].set_title('Predicted PPG')

#############################################################################
# Ground truth PPG
axes[1].plot(gt[Start_point:End_point], 'orange')
axes[1].set_title('Ground truth PPG')

plt.show()    
```

##  Heart rate (HR) 
#### "Predicted & ground truth PPG 로 부터 HR을 추정 후 차이를 비교"
- MAE (Mean absolute error) : 평균 절대값 오차
- RMSE (Root mean square error) : 평균 제곱근 오차 

```python
calculate_HR_metrics(pred, gt)
```
